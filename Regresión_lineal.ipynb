```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Create a data frame with the given data
df = pd.DataFrame({'x': [1,2,2,3,4,4,5,6], 'y': [2,3,4,4,4,6,5,7]})

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(df[['x']], df['y'])

# Predict the values of y for the given x values
predicted_y = model.predict(df[['x']])

# Create a scatter plot of the data
import matplotlib.pyplot as plt
plt.scatter(df['x'], df['y'])

# Plot the regression line
plt.plot(df['x'], predicted_y, color='pink')

# Show the plot
plt.show()
plt.show()

# Calculate the R-squared value
r_squared = model.score(df[['x']], df['y'])
print('R-squared:', r_squared)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crear el DataFrame con los datos proporcionados
data = {
    'x': [1, 2, 2, 3, 4, 4, 5, 6],
    'y': [2, 3, 4, 4, 4, 6, 5, 7]
}

df = pd.DataFrame(data)

# Variables independientes y dependientes
x = df['x']
y = df['y']

# Calcular los coeficientes de la regresión lineal
x_mean = np.mean(x)
y_mean = np.mean(y)

# Fórmulas para calcular la pendiente (b1) y el intercepto (b0)
numerator = sum((x - x_mean) * (y - y_mean))
denominator = sum((x - x_mean) ** 2)

b1 = numerator / denominator
b0 = y_mean - b1 * x_mean

# Predecir los valores
y_pred = b0 + b1 * x

# Visualizar los resultados de la regresión lineal simple
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='#6819C6', label='Datos Reales')
plt.plot(x, y_pred, color='#19C6B7', linewidth=2, label='Línea de Regresión')
plt.title('Regresión Lineal Simple')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Imprimir coeficientes
print("Coeficiente (pendiente):", b1)
print("Intercepto:", b0)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crear el DataFrame con los datos proporcionados
data = {
    'x': [1, 2, 2, 3, 4, 4, 5, 6],
    'y': [2, 3, 4, 4, 4, 6, 5, 7]
}

df = pd.DataFrame(data)

# Crear una nueva característica x^2
df['x2'] = df['x'] ** 2

# Variables independientes y dependientes
X = df[['x', 'x2']]
y = df['y']

# Añadir una columna de unos para el término de intercepto
X = np.column_stack((np.ones(len(X)), X))

# Calcular los coeficientes de la regresión lineal múltiple manualmente usando la fórmula de mínimos cuadrados
# β = (X^T * X)^(-1) * X^T * y
Xt = X.T
beta = np.linalg.inv(Xt.dot(X)).dot(Xt).dot(y)

# Coeficientes
b0 = beta[0]
b1 = beta[1]
b2 = beta[2]

# Predecir los valores
y_pred = X.dot(beta)

# Visualizar los resultados de la regresión lineal múltiple
plt.figure(figsize=(10, 6))
plt.scatter(df['x'], y, color='#6819C6', label='Datos Reales')
plt.plot(df['x'], y_pred, color='#19C6B7', linewidth=2, label='Línea de Regresión')
plt.title('Regresión Lineal Múltiple')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)  # Agregar la grilla
plt.show()

# Imprimir coeficientes
print("Coeficiente (intercepto):", b0)
print("Coeficiente (x):", b1)
print("Coeficiente (x^2):", b2)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Crear el DataFrame con los datos proporcionados
data = {
    'x': [1, 2, 2, 3, 4, 4, 5, 6],
    'y': [2, 3, 4, 4, 4, 6, 5, 7]
}

df = pd.DataFrame(data)

# Nueva característica x^2
df['x2'] = df['x'] ** 2

# Variables independientes y dependientes
X = df[['x', 'x2']]
y = df['y']

# Añadir una columna de unos para el término de intercepto
X = np.column_stack((np.ones(len(X)), X))

# Calcular los coeficientes con la fórmula de mínimos cuadrados
# β = (X^T * X)^(-1) * X^T * y
Xt = X.T
beta = np.linalg.inv(Xt.dot(X)).dot(Xt).dot(y)

# Coeficientes
b0 = beta[0]
b1 = beta[1]
b2 = beta[2]

# Predecir los valores
y_pred = X.dot(beta)

# Agregar las columnas Ypred, Error, Error^2 al DataFrame
df['Ypred'] = y_pred
df['Error'] = df['y'] - df['Ypred']
df['Error^2'] = df['Error'] ** 2

# MSE : Promedio de Error^2.
# Sr Suma de los Cuadrados Residuales: Suma de Error^2.
# RMSE Root Mean Squared Error: Raíz cuadrada de MSE.
# MAE Mean Absolute Error : Promedio valor absoluto de Error.


# Calcular MSE
mse = df['Error^2'].mean()

# Calcular Sr
sr = df['Error^2'].sum()

# Calcular RMSE
rmse = np.sqrt(mse)

# Calcular MAE
mae = df['Error'].abs().mean()



# Imprimir coeficientes, MSE, RMSE, MAE y R^2
print("Coeficiente (intercepto):", b0)
print("Coeficiente (x):", b1)
print("Coeficiente (x^2):", b2)
print("MSE:", mse)
print("Sr (sum of squared residuals):", sr)
print("RMSE:", rmse)
print("MAE:", mae)

# Mostrar DataFrame con las nuevas columnas
print("\nTabla de Resultados:")
print(df)
```
